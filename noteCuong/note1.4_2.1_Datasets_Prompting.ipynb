Bộ dữ liệu **PhoATIS** được xây dựng dựa trên chủ đề **đặt vé máy bay** và các hội thoại liên quan đến việc xử lý ngôn ngữ tự nhiên trong lĩnh vực **hệ thống hội thoại định hướng nhiệm vụ (task-oriented dialogue systems)**. Cụ thể:

### **Nội dung của PhoATIS**
1. **Chủ đề chính**:
   - Bộ dữ liệu tập trung vào các hội thoại liên quan đến việc **đặt vé máy bay**, bao gồm:
     - Tìm kiếm chuyến bay theo thời gian, địa điểm, và hãng hàng không.
     - Yêu cầu thông tin về giá vé, thời gian khởi hành/hạ cánh.
     - Hỏi đáp thông tin về các sân bay hoặc thành phố.
     - Xác nhận hoặc hủy đặt vé.

2. **Mục tiêu của bộ dữ liệu**:
   - **Intent detection (nhận diện ý định)**:
     - Phát hiện mục đích của người dùng từ các câu hội thoại. Ví dụ:
       - Ý định “tìm chuyến bay” (find_flight).
       - Ý định “hỏi giá vé” (airfare).
   - **Slot filling (điền khe)**:
     - Trích xuất các thông tin cần thiết từ câu nói của người dùng để hoàn thành nhiệm vụ.
     - Ví dụ:
       - Từ câu “Cho tôi biết các chuyến bay từ Hà Nội đến Đà Nẵng”, hệ thống cần nhận diện các khe:
         - **fromloc.city_name**: Hà Nội.
         - **toloc.city_name**: Đà Nẵng.

---

### **Tại sao PhoATIS đặc biệt?**
1. **Ngữ cảnh Việt Nam**:
   - Dữ liệu được dịch từ tiếng Anh sang tiếng Việt với các chỉnh sửa phù hợp với ngữ cảnh thực tế tại Việt Nam.
   - Thay thế các địa danh, hãng hàng không và tình huống đặc trưng ở Mỹ bằng các thông tin phù hợp với người dùng Việt Nam.

2. **Phân loại chi tiết**:
   - Gồm **28 nhãn ý định (intent labels)** như `find_flight`, `airfare`, `flight_time`, `ground_service`, v.v.
   - Gồm **82 loại khe (slot types)** như `fromloc.city_name`, `toloc.city_name`, `departure_time.time`, `airline_name`, v.v.

3. **Ứng dụng thực tế**:
   - Phục vụ nghiên cứu về **xử lý ngôn ngữ tự nhiên** (NLP) cho tiếng Việt.
   - Hỗ trợ phát triển các hệ thống chatbot hoặc trợ lý ảo định hướng nhiệm vụ trong tiếng Việt.

---

### **Một số ví dụ trong PhoATIS**
| Câu hội thoại                           | Ý định (Intent) | Các khe (Slots)                        |
|-----------------------------------------|-----------------|----------------------------------------|
| "Tìm chuyến bay từ Hà Nội đến Đà Nẵng." | `find_flight`   | `fromloc.city_name`: Hà Nội, `toloc.city_name`: Đà Nẵng |
| "Giá vé từ TP.HCM đến Hà Nội là bao nhiêu?" | `airfare`       | `fromloc.city_name`: TP.HCM, `toloc.city_name`: Hà Nội |
| "Chuyến bay nào rời sân bay Nội Bài lúc 8h sáng?" | `flight_time`   | `fromloc.airport_name`: Nội Bài, `departure_time.time`: 8h sáng |

---

PhoATIS rất hữu ích nếu bạn đang nghiên cứu hoặc xây dựng các ứng dụng NLP cho tiếng Việt, đặc biệt trong lĩnh vực hệ thống hội thoại hoặc trợ lý ảo định hướng nhiệm vụ.

## Idea 1: gen ra bộ JSON:  'question': , 'answer', 'intent' cho các cụm và các dạng bài học. 
### Prompt: (Web Version) 
```
- Với cụm từ 'go to school' gen ra 6 cặp câu hỏi câu trả lời 

dạng JSON: 'question': , 'answer', 'intent'

- question: Robot nói: tiếng việt, tiếng anh A2. Robot dạy học sinh nói cụm 'go to school'.
- answer: Học sinh trả lời nửa anh nửa việt. 
- intent: intent_positive, intent_negative, intent_neutral, intent_learn_more, intent_fallback, cant_hear. 

Note:
- Xưng hô : cậu -tớ,
- lần lượt với từng intent, các bộ question - answer - intent là trong 1 conversation không đứt đoạn
```

### **Danh sách các bài học (Flow bài học)**

1. **Điền từ vào chỗ trống (Fill in the blanks)**  
   - Ví dụ: Điền từ vào câu "Can I ________ with you, please?"  
   - Kỹ năng: Từ vựng, cấu trúc câu cơ bản.  

2. **Chọn đáp án đúng (Multiple choice)**  
   - Ví dụ: "Cậu làm gì ngay sau khi thức dậy? A) Brush teeth, B) Go to bed, C) Eat dinner."  
   - Kỹ năng: Đọc hiểu, chọn lựa từ phù hợp.

3. **Hỏi từ trong một chủ đề (Word listing by topic)**  
   - Ví dụ: Kể tên nhiều từ bắt đầu bằng "play" (play football, play games, play piano).  
   - Kỹ năng: Mở rộng vốn từ theo chủ đề.  

4. **Sắp xếp từ thành câu đúng (Rearrange words)**  
   - Ví dụ: "Games playing I video love." -> "I love playing video games!"  
   - Kỹ năng: Ngữ pháp, trật tự từ.

5. **Mở rộng hội thoại theo câu trả lời của user (Dynamic conversation)**  
   - Ví dụ: "Cậu muốn làm gì tiếp theo? A) Eat breakfast, B) Take a shower."  
   - Kỹ năng: Hội thoại, logic tình huống.

6. **Điền từ hoàn chỉnh đoạn văn (Complete the paragraph)**  
   - Ví dụ: "I wake up at ________. Then, I ________."  
   - Kỹ năng: Viết và sắp xếp ý tưởng.  

7. **Học từ chunk -> câu -> đoạn (Progressive learning)**  
   - Ví dụ: "Wake up -> I wake up at 7 am -> In the morning, I wake up at 7 am. After that, I..."  
   - Kỹ năng: Xây dựng câu từ đơn giản đến phức tạp.  

8. **Dạy từ theo sở thích của user (Personalized learning)**  
   - Ví dụ: "Cậu thích ăn gì?" -> "Cơm rang, bún, phở." -> "Cơm rang là 'fried rice' trong tiếng Anh."  
   - Kỹ năng: Từ vựng liên quan đến sở thích.  

9. **Giải thích nghĩa từ (Explain word meanings)**  
   - Ví dụ: "Family nghĩa là 'gia đình', gồm bố mẹ và anh chị em."  
   - Kỹ năng: Hiểu nghĩa và ngữ cảnh sử dụng.

**Ví dụ từng nhãn**
```
- **user_intent_positive**:  only 2 pairs  
     - User expresses satisfaction or optimism with **emotional reactions**.  
     - **Examples:**  
       - "Hay quá!", "Wow, dễ hiểu thật!", "Cảm ơn nhé, tớ hiểu rồi!", "Ồ, tuyệt thế, tớ học thêm được nhiều lắm!"  
       - "Hihi, đúng là dễ hiểu hơn rồi!", "Thật tuyệt, tớ hiểu nhanh luôn!"  
       - "Ôi, hay ghê!", "Tớ thấy cách này thú vị thật đấy!"  

  - **user_intent_negative**:  only 2 pairs  
     - User shows dislike, dissatisfaction, or complaints.  
     - **Examples:**  
       - "Khó quá, tớ không học nổi đâu!", "Cách này phức tạp quá, tớ không thích!"  
       - "Tớ không muốn học nữa!", "Học mà không nhớ nổi gì cả, chán thật!"  
       - "Tớ thấy cách này không hiệu quả chút nào!", "Cụm này chán lắm, tớ không thích!"  

  - **user_intent_neutral**:  only 2 pairs  
     - User responds indifferently or without strong emotion.  
     - **Examples:**  
       - "Ok.", "Ừ thì chắc là đúng.", "Cũng được.", "Ừm, tớ cũng không rõ nữa."  
       - "Thì cũng tạm ổn thôi.", "Chắc là thế."  
       - "Không quan trọng lắm.", "Ừ thì tớ đoán là đúng rồi."  

  - **user_intent_learn_more**:  only 2 pairs  
     - User shows curiosity or interest in learning more.  
     - **Examples:**  
       - "Chỉ tớ thêm với!", "Tớ muốn học tiếp!", "Có gì hay nữa không?"  
       - "Còn cách dùng nào khác không nhỉ?", "Tớ có thể dùng cụm này trong ngữ cảnh khác không?"  
       - "Cậu dạy tớ thêm cách dùng từ này đi!", "Tớ muốn hiểu sâu hơn, giải thích thêm nhé?"  

  - **user_intent_fallback**:  only 2 pairs  
     - User says something **unrelated or off-topic**.  
     - **Examples:**  
       - "Học cái này để làm gì nhỉ?", "Tớ vừa mua giày mới, đẹp không cậu?"  
       - "Tớ đang nghĩ xem tối nay xem phim gì.", "Ờ... mà trời hôm nay đẹp nhỉ?"  
       - "Tớ chán học rồi, chơi game đi?", "À mà cậu biết không, hôm nay tớ vừa gặp bạn cũ!"  

  - **user_silence**:  
     - User remains silent or does not respond.  
     - **Examples:**  
       - "...", "", `none`  
       - (Không có phản hồi, học sinh giữ im lặng)  
       - (Học sinh không trả lời trong thời gian dài)  
```
